### For Engine Column

```python
df[['Engine_Value','Engine_Unit']]=df['Engine'].str.split(' ',expand=True)
```

```python
df[['Engine','Engine_Value','Engine_Unit']]
```

### For Power Column

```python
df[['Power_Value','Power_Unit']]=df['Power'].str.split(' ',expand=True) ## expand=True change to dataframe
```

```python
df[['Power','Power_Value','Power_Unit']]
```

#### Dropping the unneccarsy column Mileage,Engine,Powre

### Filling the categrical data firt by mode

```python
df.columns
```

```python
cat_columns = ['Location', 'Year','Fuel_Type', 'Transmission',
       'Owner_Type', 'Colour', 'Seats', 'No. of Doors','Mileage_Unit','Engine_Unit','Power_Value',]
```

```python
for i in cat_columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

#### 1. Price Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins=30)
plt.title("Car Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
2. Company-wise Average Price (Bar Chart)
```

```python
avg_price = df.groupby('Company_Name')['Price'].mean()

plt.figure(figsize=(10,5))
plt.bar(avg_price.index, avg_price.values)
plt.xticks(rotation=90)
plt.title("Average Price by Company")
plt.xlabel("Company")
plt.ylabel("Average Price")
plt.show()
```

#### 3. Fuel Type Count (Pie Chart)

```python
Fuel = df['Fuel_Type'].value_counts()
```

```python
fuel = df['Fuel_Type'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(Fuel.values, labels=Fuel.index, autopct='%1.1f%%')
plt.title("Fuel Type Distribution")
plt.show()
```
